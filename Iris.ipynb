## Step 1: Import necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split

sns.set()
```

```python
data = datasets.load_iris()
```

```python
data.keys()
```

```python
data.target_names
```

```python
np.array(data['target_names'].value_counts)
```

```python
data.data
```

```python
df = pd.DataFrame(data=data["data"],columns=data["feature_names"])
```

```python
df["target"] = data["target"]
```

```python
df["species"] = df["target"].apply(lambda x: data.target_names[x])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
col ='sepal length (cm)'
df[col].hist()
plt.suptitle(col)
plt.show()
```

```python
sns.pairplot(df,hue="species",diag_kind="kde")
```

### We can easily distinguish the Setosa species (blue data points) from the other two species on this pairplot chart. While there's some overlap between the Versicolor and Virginica species, we can still differentiate between them with high accuracy..

```python
sns.heatmap (df.drop("species", axis=1).corr())
```

### There is a strong positive correlation between the petal dimensions and the sepal length, while the sepal width has a weak or negative correlation with the other features.

```python
df.columns
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.boxplot(x=df['species'], y=df["sepal length (cm)"], data=df, ax=axes[0,0])
sns.boxplot(x=df['species'], y=df["sepal width (cm)"], data=df, ax=axes[0,1])
sns.boxplot(x=df['species'], y=df["petal length (cm)"], data=df, ax=axes[1,0])
sns.boxplot(x=df['species'], y=df["petal width (cm)"], data=df, ax=axes[1,1])
plt.tight_layout()
plt.show()
```

### The boxplots show that petal dimensions are the most effective for separating the species, while sepal dimensions have significant overlap, making them less useful for classification

```python
#sns.violinplot(x=df["species"],y=df["petal length (cm)"])
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.violinplot(x=df['species'], y=df["sepal length (cm)"], data=df, ax=axes[0,0])
sns.violinplot(x=df['species'], y=df["sepal width (cm)"], data=df, ax=axes[0,1])
sns.violinplot(x=df['species'], y=df["petal length (cm)"], data=df, ax=axes[1,0])
sns.violinplot(x=df['species'], y=df["petal width (cm)"], data=df, ax=axes[1,1])
plt.tight_layout()
plt.show()
```
